### Top 5 player combos by offensive turnover percentage (o_TOV%):

```python
import itertools
import pandas as pd

# list of player shortcodes
players = ['LL', 'MH', 'AM7', 'AM', 'KJ', 'IK', 'JL', 'ZO', 'CH', 'MS', 'AC','PR']

# store results
combo_stats = []

for p1, p2 in itertools.combinations(players, 2):
    mask = ls['lineup'].str.contains(p1) & ls['lineup'].str.contains(p2)
    df_pair = ls[mask]

    if df_pair.empty:
        continue  # skip if this combo never played together

    agg = df_pair[['minutes', 'plus_minus', 'poss_total', 'o_poss', 'd_poss', 'pts_for', 'pts_against']].sum()

    # Compute o_TOV%
    if 'o_TOV%' in df_pair.columns and agg['o_poss'] > 0:
        total_turnovers = (df_pair['o_TOV%'] * df_pair['o_poss']).sum()
        o_tov = total_turnovers / agg['o_poss']
    else:
        o_tov = float('nan')

    combo_stats.append({
        'Combo': f"{p1}-{p2}",
        'Minutes': agg['minutes'],
        'o_TOV%': o_tov
    })

# Convert to DataFrame
combo_df = pd.DataFrame(combo_stats)
combo_df['o_TOV%'] = combo_df['o_TOV%'].round(2)
combo_df['Minutes'] = combo_df['Minutes'].round(2)

# Sort by o_TOV%
combo_df = combo_df.sort_values('o_TOV%', ascending=False)

# Display top 5
print("Top 5 player combos by offensive turnover percentage (o_TOV%):")
print(combo_df.head(10))
```
